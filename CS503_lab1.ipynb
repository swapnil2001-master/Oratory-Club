
“Data Standardization” means that the range of values is “standardized to measure how many standard deviations the value is from its mean”. 
1. Standardize the data and print the standardized data in a tabular form. **(1 point)** 
2. Plot the histogram corresponding to each attribute as done above. **(4 points)** {1*4=4}
3. Note the change in the printed data and histogram due to standardization. **(1 point)** 





```
#function for standardizing the dataset  

def Standardize(data):
    mean_of_data = data.mean(axis =0)
    standard_deviation_of_data = data.std(axis = 0)
    standardizd_data = (data-mean_of_data)/(standard_deviation_of_data)
    return standardizd_data

#functions for normalizing the dataset

def normalize(data):
    min_data = data.min(axis=0)
    max_data = data.max(axis=0)
    normalizd_data =  (data -min_data)/(max_data-min_data)
    return normalizd_data
```

# <font color = '03506f'>Standarization of data : 
<font color = '#1687a7'>Data standardization is rescaling the data such that the mean of the data is 0 and the standard deviation of the data is 1 .

it is calculated by using the formula.</font>

$$ Zs = \frac{x - \mu}{\sigma}$$
where $${\mu = mean}$$ $${\sigma = variance}$$

# <font color = '03506f'>Normalization of data :
<font color = '#1687a7'>Data normalization is rescaling the data such that the max value is 1 and the min value is 0 . (rescaling all the values between 0-1).

it is calculated by using the formula.</font>

$$Zn = \frac{X - X.min}{X.max - X.min}$$

```
# Standardize the dataset 
standardized_data = Standardize(Data)                                              # passing the n-dim array of iris.data in the functiion for calcalulation of standardized data .
standardized_data = pd.DataFrame(standardized_data,columns=iris['feature_names'])  # making a dataframe of standarized data for better visualization 


# Print dataset before standardization        
print(color.blue + '\nDataset before standardization\n'+color.end)
display(iris_df)

# Print dataset after standardization  
print(color.blue + '\n\nDataset after standardization\n'+color.end)
display(standardized_data)

# For entire standardized dataset, plot 4 histograms for sepal length (cm), sepal width (cm), petal length (cm) and petal width (cm)

#Plot histogram 
print(color.blue + '\n\nRequired Histograms\n'+color.end)
for s in iris['feature_names']:
  sns.histplot(data = standardized_data[s],edgecolor = 'green')                   # plotted the graph using seaborn we could also use matplot library .
  plt.show()
  print('\n')


#for better visulization we can use the describe property .
#display(standardized_data.describe())  
#display(iris_df.describe())
```
